```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## IMPORTING FILE

```python
#import file
data = pd.read_excel('bike_sale.xlsx')
data
```

## DATA CLEANING

```python
#Check duplicate
data.duplicated()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data["Marital Status"] = data["Marital Status"].replace (["M","S"],["Married","Single"])
```

```python
data
```

```python
data["Gender"] = data["Gender"].replace (["F","M"], ["Female","Male"])
```

```python
data
```

```python
#age grouping
def age_grouping(age):
    if age<=30:
        return "Adolescent"
    elif age<=54:
        return "Middle Age"
    else:
        return "old"
```

```python
data["Age Bracket"] = data["Age"].apply(age_grouping)
data
```

```python
piv1 = np.round(data.groupby(["Gender", "Purchased Bike"])["Income"].mean().unstack(), 0)
```

```python
piv1
```

```python
piv = np.round(piv1,0)
```

```python
piv2 = data.groupby(["Commute Distance", "Purchased Bike"])["Purchased Bike"].count().unstack()
piv2
```

```python
piv3 = data.groupby(["Age Bracket", "Purchased Bike"])["Purchased Bike"].count().unstack()
piv3
```

```python
piv.plot(kind='bar', figsize=(6,4))
plt.title("Pivot 1 charts")
plt.xlabel("Gender")
plt.ylabel("Purchased Bike")
```

```python
piv2.plot(kind='line', figsize=(7,3))
plt.title("Pivot 2 charts")
plt.xlabel("Commute Distance")
plt.ylabel("Purchased Bike")
```

```python
piv3.plot(kind='bar', figsize=(7,3))
plt.title("Pivot 3 charts")
plt.xlabel("Age Bracket")
plt.ylabel("Purchased Bike")
```
